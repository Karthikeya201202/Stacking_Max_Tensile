```python
df= pd.read_excel(r"C:\Users\yedla\Downloads\Stacking_max_stress\vznd6z234n-1\Data.xlsx")
df.head()
```

```python
X=df.iloc[:,:-1]
X.head()
```

```python
y=df.iloc[:,-1:]
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=19)
```

```python
print(len(X_train),len(X_test))
```

```python
#Normalising the data using MINMAX Scaler()

scaler=MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)

#test data
X_test_scaled= scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled,columns = X_test.columns)
```

### Modelling

```python
hidden_units1 = 13
hidden_units2 = 32
hidden_units3 = 32
learning_rate = 0.01
```

```python
#import tensorflow as tf
#from tensorflow.keras.models import Sequential
#from tensorflow.keras.layers import Dense

# Create a Sequential model
model = Sequential()

# Add an input layer with 13 neurons
model.add(Dense(units=13, input_dim=13, activation='relu'))

# Add the first hidden layer with 32 neurons and ReLU activation
model.add(Dense(units=32, activation='relu'))

# Add the second hidden layer with 32 neurons and ReLU activation
model.add(Dense(units=32, activation='relu'))

# Add an output layer with 1 neuron and linear (identity) activation
model.add(Dense(units=1, activation='linear'))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')  # You can choose an appropriate optimizer and loss function

# Print a summary of the model architecture
model.summary()
```

```python
model.fit(X_train_scaled,y_train,epochs=200, validation_split=0.1)
```

```python
from keras import backend as K
```

```python
def r2_score(y_test, y_pred):
    SS_res =  K.sum(K.square( y_test-y_pred )) 
    SS_tot = K.sum(K.square( y_test - K.mean(y_test) ) ) 
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )
```

```python
pred= model.predict(X_test_scaled)
print(K.get_value(r2_score(K.constant(y_test),pred)))
```

```python
pred_train= model.predict(X_train_scaled)
print(K.get_value(r2_score(K.constant(y_train),pred_train)))
```
